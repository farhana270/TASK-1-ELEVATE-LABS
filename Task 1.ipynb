TASK 1 : Clean and prepare Raw dataset

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("C:/Users/Farhana/Downloads/archive/netflix_titles.csv")

# Step 1: Remove duplicate rows
df = df.drop_duplicates()

# Step 2: Handle missing values
# Fill missing 'director', 'cast', and 'country' with 'Unknown'
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)

# Drop rows with missing 'date_added', 'rating', or 'duration'
df.dropna(subset=['date_added', 'rating', 'duration'], inplace=True)

# Step 3: Convert 'date_added' to datetime format
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Step 4: Strip whitespace in string columns for consistency
df['rating'] = df['rating'].str.strip()
df['type'] = df['type'].str.strip()

# Optional: Save cleaned data to a new CSV
df.to_csv("netflix_titles_cleaned.csv", index=False)

print("Data cleaning complete. Cleaned dataset saved as 'netflix_titles_cleaned.csv'")
df.head()
```
